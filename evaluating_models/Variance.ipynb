```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(0)

# Parameters
num_samples = 1000  # Number of samples to generate for each sample size
max_sample_size = 100  # Maximum sample size
sample_sizes = range(1, max_sample_size + 1)  # Sample sizes from 1 to max_sample_size

# Storage for variances of sample means
variances = []

# Generate samples and calculate variance of means
for size in sample_sizes:
    # Generate 'num_samples' samples of 'size' elements
    samples = np.random.normal(loc=0, scale=1, size=(num_samples, size))
    # Compute means of each sample
    sample_means = np.mean(samples, axis=1)
    # Compute variance of these means
    variances.append(np.var(sample_means))

# Plotting the variance against sample size
plt.figure(figsize=(10, 6))
plt.plot(sample_sizes, variances, marker='o', linestyle='-', color='b')
plt.title('Variance of Sample Means vs. Sample Size', fontsize = 20)
plt.xlabel('Sample Size', fontsize = 20)
plt.ylabel('Variance of Sample Means', fontsize = 20)
plt.grid(True)
plt.show()
```
